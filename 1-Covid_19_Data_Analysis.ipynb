```python
def hospitalization (data):
    if(data['Patient addmited to regular ward (1=yes, 0=no)'] == 1):
        return 'regular ward'
    if(data['Patient addmited to semi-intensive unit (1=yes, 0=no)'] == 1):
        return 'semi-intensive care'
    if(data['Patient addmited to intensive care unit (1=yes, 0=no)'] == 1):
        return 'intensive care'
    else :
        return 'unknown'

#creation of a new column that explicitly shows in what unit of the hospital was the patient
df['status'] = df.apply(hospitalization, axis = 1)
df.head()
```

```python
#possible relation between blood samples values and the unit of the hospital where the patient is taking care of
for col in blood_columns:
    plt.figure()
    for cat in df['status'].unique():
        sns.histplot(df[df['status'] == cat][col], label = cat, kde=True, stat = 'density')
    plt.legend()
```

#### Patients in intensive care seem to have some differences in their blood samples (e.g.: Platelets, Leukocytes), compared to ohter patients

### First quick test: Do Positive and Negative patients have the same blood values ? 

```python
print(positive_df.shape, negative_df.shape)
```

```python
#balancing the number of positive and negative patients
sampled_negative_df = negative_df.sample(positive_df.shape[0])
```

```python
from scipy.stats import ttest_ind
```

```python
def t_test(col):
    alpha = 0.02
    stats, p = ttest_ind(sampled_negative_df[col].dropna(), positive_df[col].dropna())
    if p < alpha:
        return 'Hypothesis rejected'
    else :
        return 0
```

```python
for col in blood_columns:
    print(f'{col:-<50}{t_test(col)}')
```

### First Conclusion: it seems that Postive and Negative patients DON'T have the same blood sample values.
